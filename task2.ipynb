Ignoring the records with inconsistent data

```python
r=pd.read_csv('/content/drive/MyDrive/twitter_training.csv',encoding='latin1',error_bad_lines=False)
```

```python
p=r[r['Opinion']=='Positive'].shape[0]
q=r[r['Opinion']=='Negative'].shape[0]
s=r[r['Opinion']=='Irrelevant'].shape[0]
t=r[r['Opinion']=='Neutral'].shape[0]
arr=[p,q,t,s]
arr1=['Positive','Negative','Neutral','Irrelevant']
```

```python
plt.scatter(arr1,arr,color=['green', 'red', 'blue', 'black'])
plt.title('Sentiment Pattern in Twitter')
plt.xlabel('Opinion')
plt.ylabel('Count')
plt.show()
```

```python
plt.bar(arr1,arr,color=['green', 'red', 'blue', 'black'])
plt.title('Sentiment Pattern in Twitter')
plt.xlabel('Opinion')
plt.ylabel('Count')
plt.show()
```
